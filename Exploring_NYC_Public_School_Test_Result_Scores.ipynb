```python
'''Performing exploratory data analysis on the NYC Public School Test Result'''

# Importing pandas Libraries
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

```python
''' Q.] Which NYC schools have the best math results ?
(The best math results are at least 80% of the 
*maximum possible score of 800* for math)'''

best_math_score = 80*8   # 80% of 800

# Selecting Schools and their math marks from which average math marks are greater than or equal to best_math_score
math_schools = schools[schools["average_math"] >= best_math_score][["school_name", "average_math"]]

best_math_schools = math_schools.sort_values("average_math", ascending=False)

best_math_schools
```

```python
'''Q.] What are the top 10 performing schools based on the combined SAT scores ?'''
#total_SAT = "average_math" + "average_reading" + "average_writing"

# Add Column total_SAT
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]

top_schools = schools[["school_name", "total_SAT"]]

# top 10 performing schools
top_10_schools = top_schools.sort_values("total_SAT", ascending = False).head(10)

top_10_schools
```

```python
'''Q.] Which single borough has the largest standard deviation in the combined SAT score?'''
import numpy as np

# grouping borough on total_SAT mean & std deviation values
group_borough = round(schools.groupby("borough")["total_SAT"].agg([np.mean, np.std]), 2).reset_index()
group_borough.columns = ["borough", "average_SAT", "std_SAT"]

# Selecting largest std deviation form group_borough
largest_std_dev = group_borough[group_borough["std_SAT"] == group_borough["std_SAT"].max()]

# Counting school name in each borough
numb_schools = schools.groupby("borough")["school_name"].count().reset_index()

# Creating new column for number of school in borough which has largest standard deviation in the combined SAT score
largest_std_dev["num_schools"] = numb_schools["school_name"]

largest_std_dev
```
