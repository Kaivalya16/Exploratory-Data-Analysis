```python
'''Performing exploratory data analysis on the netflix_data.csv data 
to understand more about movies from the 1990s decade.'''

# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
''' Q.] What was the most frequent movie duration in the 1990s ? '''

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
# print(netflix_df)

# Filtering movies from year 1990 to 1999
import numpy as np
filter = np.logical_and(netflix_df['release_year'] >= 1990 , netflix_df['release_year'] < 2000)
decade_90s = netflix_df[filter]

# Visualizing distribution of durations in Histogram
plt.hist(decade_90s['duration'], bins = 15) 
plt.show()

duration = 100
```

```python
''' Q.] Count the number of short action movies released in the 1990s. '''

# Action movies in 90s
action_movies = decade_90s[decade_90s['genre'] == 'Action']
short_movie_count = 0

# Iterating through action movies
for label, row in action_movies.iterrows():
    # Looking for short movies
    if row['duration'] < 90:
        short_movie_count += 1

short_movie_count
```
